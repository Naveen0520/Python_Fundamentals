```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x=[1,2,3]
y=[1,4,9]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.title('sales information')
plt.xlabel('sales')
plt.ylabel('revenue')
plt.plot(x,y)
plt.show()
```
